```python
from datasets import load_from_disk, Audio
import os
from dsp import plot_save_mel_spectrogram
import numpy as np
from IPython import display as disp
```

```python
ds = load_from_disk('../datasets/preprocessed_subset')

for example in ds.select(range(25,30)):
    audio = example['audio']
    print(example['time_freq_bounds'])
    plot_save_mel_spectrogram(np.array(audio['array']), audio['sampling_rate'], details=example['scientific_name'])
    disp.display(disp.Audio(audio['array'], rate=audio['sampling_rate']))
```
